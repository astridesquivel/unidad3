Plotting sin linea.
Para trasar solo los marcadores,puedes usar notacion de cadena de acceso directo 'o',que significa "anillos".
Ejemplo:Dibuja dos puntos en el diagrama,uno en la posicion(1,3) y el otro en la posicion(8,10)

```python
import matplotlib.pyplot as plt
import numpy as np

puntosx = np.array([1,8])
puntosy = np.array([3,10])

plt.plot(puntosx,puntosy,'o')
plt.show()
```

Diagramas de dispersion (scatter)

se emplean para demostrar la relacion entre dos variables numericas,detectando correlaciones o agrupamientos,
Ejemplo

```python
import matplotlib.pyplot as plt

altura = [150,160,170,180,190,200]
peso = [50,60,65,80,85,95]

plt.scatter(altura,peso,color = "red",marker= "x")
plt.title("Relacion entre la altura y el peso")
plt.xlabel("Altura (cms)")
plt.ylabel("peso(kg)")
plt.show()
```

Graficos de barra (bar y barh)
Sirven para comparar categorias o variables tanto en orientacion vertical como horizontal.
Ejemplo de barras vesticales:

```python
import matplotlib.pyplot as plt

categorias = ["A","B","C","D"]
valores =[4,7,1,8]

plt.bar(categorias,valores,color = "orange")
plt.title("graficas de barras")
plt.xlabel("categoria")
plt.ylabel("valores")
plt.show()
```
