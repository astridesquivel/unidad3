ACTIVIDAD 3 conceptos basicos de la libreria Matplotlib

Matplotlib es una libreria de trasado de graficos de bajo nivel en python que sirve como una herramienta de visualizacion,matplotlib es de codigo abierto y podemos utilizarlo libremente.Matplotlib esta escrito principalmente en python con algunos segmentos escritos en C.

Instalacion de matplotlib,si tiene python y pip ya instalado en tu sistema,instala matplotlib usando el siguinte comando:
pip install matplotlib
desde la linea de comandos.
Importar matplotlib

Una vez instalado,importalo en tus aplicaciones agregando el modulo "import",ejemplo:
import matplotlib
Matplotlib Pyplot
pyplot la mayoria de las herramientas de matplotlib pyplot se encuentra dentro de su modulo pyplot se importa bajo el alias "PLT".

import matplotlib.pyplot as plt
ahora el paquete pyplot puede dedominarce el PLT
